# Parte 2

En esta notebook se describe el procesamiento de imagen de una captura de un camino.

La correcta implementación de procesamiento de imagen permitiría tener modelos de conducción autónoma con adquisición de datos eficaz.

## 0. Librerías básicas

Librerías básicas de esta notebook

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

### 1. Abrir la imagen para procesarla

```python
plt.figure(figsize=(14,14))
img = cv.imread("figuras/camino.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

## 2. Conversión a escala de grises

```python
img = cv.imread("figuras/camino.jpg")
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
plt.show()
```

## 3. Filtros

### 3.1 Filtro Guassiano

```python
img = cv.imread("figuras/camino.jpg")
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
plt.imshow(img_gauss, cmap="gray")
plt.show()
```

## 4. Thresholding

```python
img = cv.imread("figuras/camino.jpg")
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray, (3,3), 0)
thr, img_thr = cv.threshold(img_gauss, 100, 255, cv.THRESH_BINARY)
plt.imshow(img_thr, cmap="gray")
plt.show()
```

## 5. Redimensionando

```python
img = cv.imread("figuras/camino.jpg")
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray, (3,3), 0)
thr, img_thr = cv.threshold(img_gauss, 100, 255, cv.THRESH_BINARY)
altura = img.shape[0]
ancho = img.shape[1]
ratio = 0.08
img_r = cv.resize(img_thr, (int(ancho*ratio),int(altura*ratio)), interpolation=cv.INTER_NEAREST)
print("Altura original: ", altura, "px")
print("Ancho original: ", ancho, "px")
print("")
print("Altura resized: ", altura*ratio, "px")
print("Ancho resized: ", ancho*ratio, "px")
plt.imshow(img_r, cmap="gray")
plt.show()
```

## 6. Todas las operaciones

Procesando la imagen

```python
img = cv.imread("figuras/camino.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr = cv.threshold(img_gauss, 100, 255, cv.THRESH_BINARY)
alto = img.shape[0]
ancho = img.shape[1]
ratio = 0.08
img_r = cv.resize(img_thr, (int(ancho*ratio), int(alto*ratio)), interpolation=cv.INTER_NEAREST)
```

Mostrando resultados

```python
plt.figure(figsize=(20,12))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("Imagen original")
plt.subplot(2,3,2)
plt.imshow(img_gray, cmap="gray")
plt.title("Imagen con escala de grises")
plt.subplot(2,3,3)
plt.imshow(img_gauss, cmap="gray")
plt.title("Imagen con filtro Gaussiano")
plt.subplot(2,3,4)
plt.imshow(img_thr, cmap="gray")
plt.title("Thresholding")
plt.subplot(2,3,5)
plt.imshow(img_r, cmap="gray")
plt.title("Imagen con ratio = 0.08")
plt.show()
```
